# Creating a 3D Chart

```python
def f(x,y):
    r = 3**(-x**2-y**2)
    return 1/(r+1)
```

```python
x_1 = np.linspace(-2,2,200)
y_1 = np.linspace(-2,2,200)
x_1, y_1 = np.meshgrid(x_1, y_1)
```

```python
fig = plt.figure(figsize=(16,12))
axes = fig.gca(projection='3d')

axes.set_xlabel('X', fontsize = 20)
axes.set_ylabel('Y', fontsize = 20)
axes.set_zlabel('f(X,Y)', fontsize = 20)

axes.plot_surface(x_1,y_1,f(x_1,y_1), cmap=cm.coolwarm, alpha = 0.4)
```
