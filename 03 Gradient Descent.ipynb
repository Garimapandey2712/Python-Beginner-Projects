```python
def df(x):
    return 2*x + 1
```

## Gradient Descent

```python
new_x= 3.5
previous_x= 0
step_multiplier= 0.1
precision= 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(100):
    previous_x = new_x
    gradient = df(new_x)
    new_x = previous_x - gradient*step_multiplier
    close = abs(new_x - previous_x)
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    if close < precision:
        break
    
print('minimum at x:', new_x)
print('slope at this point:', df(new_x))
print('f(x) at this point:', f(new_x))
print(n)
```
